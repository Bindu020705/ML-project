```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("smartcart_customers.csv")
df
```

```python
df.isnull().sum()
```

# Data preprocessing

```python
# handle missing values
df["Income"] = df["Income"].fillna(df["Income"].median())
```

```python
df.head()
```

```python
df.isnull().sum()
```

# feature Engineering

# addFeature

```python
df["Age"] = 2026-df["Year_Birth"]
```

```python
df
```

```python
#customer date chaged
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"],dayfirst = True)
refernece_date = df["Dt_Customer"].max()
df["cutomer_dates"] = (refernece_date - df["Dt_Customer"]).dt.days
```

```python
df
```

```python
df.columns
```

```python
#spending 
df["total_spending"] = df["MntWines"]+df["MntFruits"]+df["MntMeatProducts"]+df["MntFishProducts"]+df["MntSweetProducts"]+df["MntGoldProds"]
```

```python
df["total_child"] = df["Kidhome"]+df["Teenhome"]
```

```python
df
```

```python
#education

df["Education"].value_counts()
df["Education"] = df["Education"].replace({
    "Basic":"Undergraduate","2n Cycle":"Undergraduate",
    "Graduation":"Graduate","Master":"Postgraduate","PhD":"Postgraduate"
})
#undergrad,grad,postgrad
df["Education"].value_counts()
```

```python
#martial status data
df["Marital_Status"].value_counts()
```

```python
df["living_with"] = df["Marital_Status"].replace({
    "Married":"partner","Together":"partner","Single":"Alone",
    "Divorced":"Alone","Widow":"Alone","Absurd":"Alone","YOLO":"Alone"
})
```

```python
df["living_with"].value_counts()
```

# drop unecessary columns

```python
#drop
cols = ["ID","Year_Birth","Marital_Status","Kidhome","Teenhome","Dt_Customer"]
spending = ["MntWines","MntFruits","MntMeatProducts","MntFishProducts","MntSweetProducts","MntGoldProds"]
col_to_drop = cols+spending
df_cleaned = df.drop(columns = col_to_drop)
```

```python
df_cleaned.shape
```

# outliers

```python
col = ["Income","Recency","Response","Age","total_spending","total_child"]
#relative plots of col
sns.pairplot(df_cleaned[col])
```

# encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
cat_col = ["Education","living_with"]
enc_col = ohe.fit_transform(df_cleaned[cat_col])

enc_df = pd.DataFrame(
    enc_col.toarray(), 
    columns=ohe.get_feature_names_out(cat_col), 
    index=df_cleaned.index
)

df_encod = pd.concat([df_cleaned.drop(columns=cat_col), enc_df], axis=1)
```

```python
df_encod.head()
```

# scaling

```python
from sklearn.preprocessing import StandardScaler

X = df_encod
```

```python
scaler =  StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Visualization

```python
X_scaled.shape
```

```python
#2d
from sklearn.decomposition import PCA
pca = PCA(n_components = 3)

X_pca = pca.fit_transform(X_scaled)
```

```python
sns.scatterplot(x = X_pca[:,0],y=X_pca[:,1])
```

```python
# 3d-plot
fig = plt.figure(figsize=(8, 6))

ax = fig.add_subplot(111, projection="3d")

ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2])
```

# Analyize K value

## 1.Elbow method

```python
from sklearn.cluster import KMeans
from kneed import KneeLocator

wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters = k,random_state=42)
    kmeans.fit_predict(X_pca)
    wcss.append(kmeans.inertia_)
```

```python
knee = KneeLocator(range(1,11),wcss,curve = "convex",direction = "decreasing")
optimal_k = knee.elbow
```

```python
print("best k = ",optimal_k)
```

```python
plt.plot(range(1,11),wcss,marker="o")
```

## silihouette score

```python
from sklearn.metrics import silhouette_score

scores=[]
for k in range(2,11):
    kmeans = KMeans(n_clusters = k,random_state=42)
    labels = kmeans.fit_predict(X_pca)
    score = silhouette_score(X_pca,labels)
    scores.append(score)
```

```python
plt.plot(range(2,11),scores,marker="o")
```

# combine for kmeans,silhouette

```python
k_range = range(2,11)

fig,ax1 = plt.subplots(figsize = (8,6))
ax1.plot(k_range,wcss[:len(k_range)],marker = "o",color = "blue")
ax2=ax1.twinx()
ax2.plot(k_range,scores[:len(k_range)],marker = "x",color = "red")
```

# Clustering

```python
# k_means
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 4,random_state=42)
labels_kmeans = kmeans.fit_predict(X_pca)
```

```python
# 3d-plot
fig = plt.figure(figsize=(8, 6))

ax = fig.add_subplot(111, projection="3d")

ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2],c=labels_kmeans)
```

```python
# Agglometrative clustering

from sklearn.cluster import AgglomerativeClustering
agg = AgglomerativeClustering(n_clusters=4,linkage = "ward")
labels_agg = agg.fit_predict(X_pca)
```

```python
# 3d-plot
fig = plt.figure(figsize=(8, 6))

ax = fig.add_subplot(111, projection="3d")

ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2],c=labels_agg)
```

# characterization of clusters

```python
df_cleaned.drop("labels",axis = 1)
df_cleaned["cluster"] = labels_agg
```
